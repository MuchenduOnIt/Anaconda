```python
#importing pandas module
import pandas as pd
import matplotlib.pyplot as plt
steve = pd.read_csv("data.csv")
steve.head()
steve.tail()
```

```python
#drawing histogram
import matplotlib.pyplot as plt
plt.hist(steve.AGE)
plt.title("Histogram")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend("Key")
plt.show()
```

```python
#drawing scatterplot
import matplotlib.pyplot as plt
plt.scatter(steve.AGE,steve.HEIGHT)
plt.title("Scatter")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend("Key")
plt.show()
```

```python
#drawing piechart
import matplotlib.pyplot as plt
plt.pie(steve.SLEEP)
plt.title("Pie chart")
plt.legend("Key")
plt.show()
```
